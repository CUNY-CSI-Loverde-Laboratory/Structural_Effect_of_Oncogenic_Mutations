### This code plots the output files from Normal mode analysis (NMA) generated from Bio3d. NMA was done on the alpha carbons (CùõÇ) of histone H4 and H2B. The fluctuation of the first three modes were combined in this analysis. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Import output files from NMA
H3_WT_015M=pd.read_csv('H3_WT_015M_m79.csv', delimiter=',')
H4_WT_015M=pd.read_csv('H4_WT_015M_m79.csv', delimiter=',')
H2A_WT_015M=pd.read_csv('H2A_WT_015M_m79.csv', delimiter=',')
H2B_WT_015M=pd.read_csv('H2B_WT_015M_m79.csv', delimiter=',')

H3_E76K_015M=pd.read_csv('H3_E76K_015M_m79.csv', delimiter=',')
H4_E76K_015M=pd.read_csv('H4_E76K_015M_m79.csv', delimiter=',')
H2A_E76K_015M=pd.read_csv('H2A_E76K_015M_m79.csv', delimiter=',')
H2B_E76K_015M=pd.read_csv('H2B_E76K_015M_m79.csv', delimiter=',')

H3_R92T_015M=pd.read_csv('H3_R92T_015M_m79.csv', delimiter=',')
H4_R92T_015M=pd.read_csv('H4_R92T_015M_m79.csv', delimiter=',')
H2A_R92T_015M=pd.read_csv('H2A_R92T_015M_m79.csv', delimiter=',')
H2B_R92T_015M=pd.read_csv('H2B_R92T_015M_m79.csv', delimiter=',')
```

```python
# Plot of the fluctuations against the residues of each histone subunits

fig, axs = plt.subplots(4, figsize=(8, 12), sharey=False, sharex=False,) 
axs[0].plot(H3_WT_015M['x'], color = "navy", label="WT")
axs[0].plot(H3_E76K_015M['x'], color = "red", label="E76K")
axs[0].plot(H3_R92T_015M['x'], color = "green", label="R92T")
axs[0].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[0].set_xlabel(r"$\bf{\it{Histone\ H3}}$", fontsize = 18.0, fontweight="bold")
axs[0].legend(loc='upper right', fontsize='large',prop={'weight':'bold'})
axs[0].tick_params(labelsize=15, pad=1)


axs[1].plot(H4_WT_015M['x'], color = "navy", label="WT")
axs[1].plot(H4_E76K_015M['x'], color = "red", label="E76K")
axs[1].plot(H4_R92T_015M['x'], color = "green", label="R92T")
axs[1].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[1].set_xlabel(r"$\bf{\it{Histone\ H4}}$", fontsize = 18.0, fontweight="bold")
axs[1].legend(loc='upper right', fontsize='large',prop={'weight':'bold'})
axs[1].tick_params(labelsize=15, pad=1)

axs[2].plot(H2A_WT_015M['x'], color = "navy", label="WT")
axs[2].plot(H2A_E76K_015M['x'], color = "red", label="E76K")
axs[2].plot(H2A_R92T_015M['x'], color = "green", label="R92T")
axs[2].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[2].set_xlabel(r"$\bf{\it{Histone\ H2A}}$", fontsize = 18.0, fontweight="bold")
axs[2].legend(loc='upper center', fontsize='large',prop={'weight':'bold'})
axs[2].tick_params(labelsize=15, pad=1)

axs[3].plot(H2B_WT_015M['x'], color = "navy", label="WT")
axs[3].plot(H2B_E76K_015M['x'], color = "red", label="E76K")
axs[3].plot(H2B_R92T_015M['x'], color = "green", label="R92T")
axs[3].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[3].set_xlabel(r"$\bf{\it{Histone\ H2B}}$", fontsize = 18.0, fontweight="bold")
axs[3].legend(loc='upper right', fontsize='large',prop={'weight':'bold'})
axs[3].tick_params(labelsize=15, pad=1)

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.tight_layout()
# show plot
plt.show()
```

```python
# Plot of the fluctuations of the histone core (tails excluded) against the residues for histone H4 and H2B subunits

fig, axs = plt.subplots(2, figsize=(8, 6), sharey=False, sharex=False,) 

axs[0].plot(H4_WT_015M['x'], color = "navy", label="WT")
axs[0].plot(H4_E76K_015M['x'], color = "red", label="E76K")
axs[0].plot(H4_R92T_015M['x'], color = "green", label="R92T")
axs[0].set_xlim(20,102)
axs[0].set_ylim(0,0.3)
axs[0].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[0].set_xlabel(r"$\bf{\it{Histone\ H4}}$", fontsize = 18.0, fontweight="bold")
axs[0].legend(loc='upper right', fontsize='large',prop={'weight':'bold'})
axs[0].tick_params(labelsize=15, pad=1)
axs[0].spines["top"].set_linewidth(3)
axs[0].spines["left"].set_linewidth(3)
axs[0].spines["right"].set_linewidth(3)
axs[0].spines["bottom"].set_linewidth(3)


axs[1].plot(H2B_WT_015M['x'], color = "navy", label="WT")
axs[1].plot(H2B_E76K_015M['x'], color = "red", label="E76K")
axs[1].plot(H2B_R92T_015M['x'], color = "green", label="R92T")
axs[1].set_xlim(23,122)
axs[1].set_ylim(0,1.5)
axs[1].set_ylabel(r"$\bf\it{Fluctuations}$", fontsize = 18.0, fontweight="bold")
axs[1].set_xlabel(r"$\bf{\it{Histone\ H2B}}$", fontsize = 18.0, fontweight="bold")
axs[1].legend(loc='upper right', fontsize='large',prop={'weight':'bold'})
axs[1].tick_params(labelsize=15, pad=1)
axs[1].spines["top"].set_linewidth(3)
axs[1].spines["left"].set_linewidth(3)
axs[1].spines["right"].set_linewidth(3)
axs[1].spines["bottom"].set_linewidth(3)

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.tight_layout()
# plt.xlim([0, 237])
plt.show()
```
