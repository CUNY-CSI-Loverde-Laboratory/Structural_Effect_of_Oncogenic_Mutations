### This code calculates the distance of the center of geometry of interacting amino acids. The distance of center of geometry between H2BE76 and H4R92 in the wild-type, H2BK76 and H4R92 in the H2BE76K mutant system and H2BE76 and H4T92 in the H4R92T mutant system. 

```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from MDAnalysis.analysis import distances
import seaborn as sns
import math as m
import pyblock
from numpy.linalg import norm

# Create Universes for all systems
top1= '/home/augustine/Nucleosome_system/1kx5/1kx5_0.15M/dry_1kx5_015M_OPC.prmtop'
traj1='/home/augustine/Nucleosome_system/1kx5/1kx5_0.15M/dry_1kx5_015M_6us.xtc'
u1 = mda.Universe(top1, traj1)

topology_E76K= '/home/augustine/Nucleosome_system/E76K/E76K_0.15M/dry_E76K_15M_OPC.prmtop'
trajectory_E76K='/home/augustine/Nucleosome_system/E76K/E76K_0.15M/dry_E76K_015M_6us.xtc'
u2 = mda.Universe(topology_E76K, trajectory_E76K)

topology_R92T= '/home/augustine/Nucleosome_system/R92T/R92T_1/R92T_0.15M/dry_R92T_015M_OPC.prmtop'
trajectory_R92T='/home/augustine/Nucleosome_system/R92T/R92T_1/R92T_0.15M/dry_R92T_015M_6us.xtc'
u3 = mda.Universe(topology_R92T, trajectory_R92T)
#0.15M
```

#### Wild-Type system at 0.15 M


```python
#DISTANCE BETWEEN H2BE76 AND H4R92 #

Distance_ER = []

for ts in u1.trajectory[8300:21000]:
    a=u1.select_atoms('resid 732 and name CA').center_of_geometry()
    b=u1.select_atoms('resid 521 and name CA').center_of_geometry()
    dist=(norm(a - b))
    Distance_ER.append((ts.frame, dist))
print(Distance_ER)

#Save data to csv file
DIS = pd.DataFrame(Distance_ER)
print(DIS)
DIS.to_csv("Dist_ER_1kx5_6us.csv")
```

#### H2BE76K system at 0.15 M

```python
#DISTANCE BETWEEN H2BK76 AND H4R92 #

Distance_KR = []
for ts in u2.trajectory[12000:25000]:
    a=u2.select_atoms('resid 732 and name CA').center_of_geometry()
    b=u2.select_atoms('resid 521 and name CA').center_of_geometry()
    dist=(norm(a - b))
    Distance_KR.append((ts.frame, dist))
print(Distance_KR)

#Save data to csv file
DIS1 = pd.DataFrame(Distance_KR)
print(DIS1)
DIS1.to_csv("Dist_KR_E76K_6us.csv")
```

#### H4R92T system at 0.15 M

```python
#DISTANCE BETWEEN H2BE76 AND H4T92 #

Distance_ET = []
for ts in u3.trajectory[12000:25000]:
    a=u3.select_atoms('resid 732 and name CA').center_of_geometry()
    b=u3.select_atoms('resid 521 and name CA').center_of_geometry()
    dist=(norm(a - b))
    Distance_ET.append((ts.frame, dist))
print(Distance_ET)

#Save data to csv file
DIS2 = pd.DataFrame(Distance_ET)
print(DIS2)
DIS2.to_csv("Dist_ET_R92T_6us.csv")
```

```python
# quick plot of Distance between H2BE76K and H4R92T against time for all systems at 6 𝝁s

plt.subplots(figsize=(9, 7))
x_val1 = [x[0]/3500 for x in Distance_ER]
y_val1 = [x[1] for x in Distance_ER]
x_val2 = [x[0]/4166.667 for x in Distance_KR]
y_val2 = [x[1] for x in Distance_KR]
x_val3 = [x[0]/4166.667 for x in Distance_ET]
y_val3 = [x[1] for x in Distance_ET]
plt.plot(x_val1, y_val1, lw=2, color='blue', label='WT')
plt.plot(x_val2, y_val2, lw=2, color='r', label='E76K')
plt.plot(x_val3, y_val3, lw=2, color='green', label='R92T')
plt.xlim([0, 6])
#plt.title("Distance between E76 and R92T \n over time", weight="bold")
plt.ylabel(r"${d_{E76K-R92}}$", fontsize = 20.0, fontweight="bold")
plt.xlabel("$Time(\u03BCs)$", fontsize = 20.0, fontweight="bold")
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.tick_params(labelsize=15, pad=1)
plt.tight_layout()
plt.legend(loc=(0.75,0.85), fontsize='large', title='0.15M', title_fontsize='large', prop={'weight':'bold'})
plt.show()
```
