### This code calculates the distance of the center of geometry of interacting amino acids. The distance of center of geometry between H2BY83 and H4K91 in the wild-type, H2BE76K and H4R92T mutant system. 

```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from MDAnalysis.analysis import distances
import seaborn as sns
import math as m
import pyblock
from numpy.linalg import norm

top1= '/home/augustine/Nucleosome_system/1kx5/1kx5_0.15M/dry_1kx5_015M_OPC.prmtop'
traj1='/home/augustine/Nucleosome_system/1kx5/1kx5_0.15M/dry_1kx5_015M_6us.xtc'
u1 = mda.Universe(top1, traj1)

topology_E76K= '/home/augustine/Nucleosome_system/E76K/E76K_0.15M/dry_E76K_15M_OPC.prmtop'
trajectory_E76K='/home/augustine/Nucleosome_system/E76K/E76K_0.15M/dry_E76K_015M_6us.xtc'
u2 = mda.Universe(topology_E76K, trajectory_E76K)

topology_R92T= '/home/augustine/Nucleosome_system/R92T/R92T_1/R92T_0.15M/dry_R92T_015M_OPC.prmtop'
trajectory_R92T='/home/augustine/Nucleosome_system/R92T/R92T_1/R92T_0.15M/dry_R92T_015M_6us.xtc'
u3 = mda.Universe(topology_R92T, trajectory_R92T)
# #0.15M
```

#### Wild-Type system at 0.15 M

```python
#DISTANCE BETWEEN H2BY83 AND H4K91 #
from numpy.linalg import norm
from MDAnalysis.analysis import distances

Distance_YK = []
for ts in u1.trajectory[8300:21000]:
    a=u1.select_atoms('resid 520 and name CA').center_of_geometry()
    b=u1.select_atoms('resid 739 and name CA').center_of_geometry()
    dist=(norm(a - b))
    Distance_YK.append((ts.frame, dist))
print(Distance_YK)

#Save data to csv file
DI1 = pd.DataFrame(Distance_YK)
print(DI1)
DI1.to_csv("Dist_YK_1kx5_6us.csv")
```

#### H2BE76K system at 0.15 M

```python
#DISTANCE BETWEEN H2BY83 AND H4K91 #

Distance_YK_E76K = []
for ts in u2.trajectory[12000:25000]:
    x=u2.select_atoms('resid 520 and name CA').center_of_geometry()
    y=u2.select_atoms('resid 739 and name CA').center_of_geometry()
    dist2=(norm(x - y))
    Distance_YK_E76K.append((ts.frame, dist2))
print(Distance_YK_E76K)

#Save data to csv file
DI2 = pd.DataFrame(Distance_YK_E76K)
print(DI2)
DI2.to_csv("Dist_YK_E76K_6us.csv")
```

#### H4R92T system at 0.15 M

```python
#DISTANCE BETWEEN H2BY83 AND H4K91 #

Distance_YK_R92T = []
for ts in u3.trajectory[12000:25000]:
    m=u3.select_atoms('resid 520 and name CA').center_of_geometry()
    n=u3.select_atoms('resid 739 and name CA').center_of_geometry()
    dist3=(norm(m - n))
    Distance_YK_R92T.append((ts.frame, dist3))
print(Distance_YK_R92T)

#Save data to csv file
DI3 = pd.DataFrame(Distance_YK_R92T)
print(DI3)
DI3.to_csv("Dist_YK_R92T_6us.csv")
```

```python
# plot of Distance between H2BY83 and H4K91 against time for all systems at 6 𝝁s

plt.subplots(figsize=(9, 7))
x_val3 = [x[0]/3500 for x in Distance_YK]
y_val3 = [x[1] for x in Distance_YK]
x_val4 = [x[0]/4166.667 for x in Distance_YK_E76K]
y_val4 = [x[1] for x in Distance_YK_E76K]
x_val5 = [x[0]/4166.667 for x in Distance_YK_R92T]
y_val5 = [x[1] for x in Distance_YK_R92T]
plt.plot(x_val3, y_val3, lw=2, color='blue', label='WT')
plt.plot(x_val4, y_val4, lw=2, color='r', label='E76K')
plt.plot(x_val5, y_val5, lw=2, color='green', label='R92T')
plt.xlim([0, 6])
#plt.title("Distance between Y83 and K91 \n over time", weight="bold")
plt.ylabel(r"${d_{Y83-K91}}$", fontsize = 20.0, fontweight="bold")
plt.xlabel("$Time(\u03BCs)$", fontsize = 20.0, fontweight="bold")
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.tick_params(labelsize=18, pad=1)
plt.tight_layout()
plt.legend(loc=(0.60, 0.83), fontsize='large', title='0.15M', title_fontsize='large', prop={'weight':'bold'})
plt.show()
```
